Obtaining the architecture of the ResNet instance. 

```python
def get_architecture(net):
    architecture = {}
    for name, module in net.named_modules():
        if 'input_stem' in name:
            architecture['input_stem'] = module
        elif 'blocks' in name:
            if 'blocks' not in architecture:
                architecture['blocks'] = []
            architecture['blocks'].append(module)
        elif 'classifier' in name:
            architecture['classifier'] = module
    return architecture

architecture = get_architecture(resnet)
print(architecture)
```

Automating the creation of ResNet instances using nested loops. 

```python

#Randomizing Base_Depth_List
# f1 = torch.randint(low=0, high=10, size=(1,4))
# f2 = torch.randint(low=0, high=10, size=(1,4))
f1 = random.randint(0,10)
f2 = random.randint(0,10)
BDL = [f1,f1,2*f1,f1]
#bd_list = np.array(BDL)
print(BDL)

# Defining the input stem
input_stem = [ConvLayer(in_channels=3, out_channels=64, kernel_size=3)]

# Defining the classifier
classifier = LinearLayer(ResNets.STAGE_WIDTH_LIST[-1], 1000)

# Defining the blocks
blocks = []

# Creating a list to store the ResNet models
resnet_models = []

for i, width in enumerate(ResNets.STAGE_WIDTH_LIST):
    # Define the blocks
    # blocks = []
    for j in range(BDL[i]):
        in_channels = width
        out_channels = width
        blocks.append(ResidualBlock(in_channels, out_channels))

# Creating the ResNet model
resnet_model = ResNets(input_stem, blocks, classifier)

# Appending the ResNet model to the list
#resnet_models.append(resnet_model)

#print(resnet_models[0])
print(resnet_model)

```

Training this ResNet instance on ImagNet dataset. 

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
#Defining the loss function. 
criterion = nn.CrossEntropyLoss()

#Defining the optimizer. 
optimizer = optim.SGD(resnet_model.parameters(), lr=0.01, momentum=0.9)

resnet_model.train()


for epoch in range(100):
    for i, (images, labels) in enumerate(data_loader):

        optimizer.zero_grad()

        outputs = resnet_model(images)

        loss = criterion(outputs, labels)

        loss.backward()

        optimizer.step()

    print("Epoch {}: Loss = {}".format(epoch, loss.item()))

torch.save(resnet_model.state_dict(), "resnet_model.pth")
```
